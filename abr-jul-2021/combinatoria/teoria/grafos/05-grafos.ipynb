## <span style="color:#8B0000;font-family:Papirus">Grafos</span>

Un **grafo** es un triple $G = (V(G), E(G), f_G)$, en el cual $V(G)$ y $E(G)$ son conjuntos y $f_G$ es una función que asigna a cada $e\in E(G)$ un par **no ordenado** $\{x, y\}\subset V(G)$. 

Por ejemplo, en el análisis del problema 1, $V(G) = \{a, b, c, d\}$, $E(G) = \{e_1, e_2, e_3, e_4, e_5, e_6, e_7\}$ y
$$ $$
$$f_G(e_1) = \{a, b\} = f_G(e_2)\quad f_G(e_3) = \{a, c\} \quad 
f_G(e_4) = \{a, d\} = f_G(e_5)$$
$$ $$
$$f_G(e_6) = \{d, c\} \qquad f_G(e_7) = \{b, c\}$$

<span style="color:#8B0000;font-family:Papirus">**Vocabulario:**</span> $v\in V(G) \rightarrow v$ es **vértice**. $\qquad e\in E(G) \rightarrow e$ es **lado** (o **arista**). 

Dados $v$, $w\in V(G)$, $\{v, w\} = f_G(e)$ para algún $e\in E(G)\rightarrow v$ y $w$ son **vecinos** (o **adyacentes**).

En un grafo, es posible que un vértice sea vecino de sí mismo, en ese caso el lado se llama **bucle**. Por ejemplo, el siguiente grafo tiene un bucle en el vértice $a$

El grafo $G$ es **simple** si no tiene bucles y $f_G$ es inyectiva (es decir, no tiene lados *paralelos*). En este caso, identificamos $E(G)$ con el rango de $f_G$. Es decir, los lados **son** los pares no ordenados de vértices adyacentes.

Por ejemplo, el grafo anterior y el del problema 1 no son simples. Los de los problemas 2 y 3 son simples. El del problema 4 puede no ser simple (es posible que entre dos ciudades haya más de una ruta directa). 

En el análisis del problema 2, hablamos de *triángulos* en el grafo, es decir, grafos con tres vértices que tienen la propiedad de que todos los vértices son vecinos entre sí. En general, un **clique** en un grafo es un conjunto de vértices que son adyacentes dos a dos. Así que los vértices de un triángulo forman un clique de tres elementos. Un conjunto de vértices es **independiente** si sus elementos son no adyacentes dos a dos. En estos términos, el problema 2 se expresa así:

> ¿Es cierto que en todo grafo con 6 vértices hay un clique con tres vértices o un conjunto independiente con tres vértices?

Los conceptos *clique* y *conjunto independiente* sugieren de inmediato el concepto de *complemento* de un grafo: El **complemento** de un grafo simple $G$ es el grafo simple $\bar{G}$ tal que $V(\bar{G}) = V(G)$ y $e\in E(\bar{G})$ si y sólo si $e\not \in E(G)$. En estos términos, $V(G)$ es un clique si y sólo $V(\bar{G})$ es un conjunto independiente.

Para el problema 5, nos interesan las conexiones sucesivas entre vértices. Esto siguere algunas definiciones. Un **camino** es un grafo simple $P$ cuyos vértices se pueden ordenar de manera que dos vértices son vecinos si y sólo si son consecutivos en dicho orden. Es decir, $V(P) = \{v_1, v_2, \dots, v_m\}$ y $E(P) = \{\{v_1, v_2\}, \{v_2, v_3\}, \dots, \{v_{m-1}, v_m\}\}$. Decimos que $v_1$ y $v_m$ son los **extremos** de  $P$. 

Por ejemplo, si consideramos en el grafo anterior los vértices $\{3, 5, 6, 4\}$ (en ese orden) junto con los lados $\{\{3, 5\}, \{5, 6\}, \{6, 4\}\}$, obtenemos un camino con extremos $3$, $4$.

Así que resolver el problema significa encontrar, dadas las ciudades $a$, $b$, el camino *más corto* que tiene como extremos $a$ y $b$.  

<span style="color:#8B0000;font-family:Papirus">**NOTACIÓN:**</span> En adelante, para un grafo simple $G$, si $v$ y $w$ son vértices vecinos; escribiremos $vw$ en lugar de $\{v, w\}$ para referirnos al lado correspondiente.

Un **ciclo** es un grafo $C$ con la misma cantidad de vértices que de lados, cuyos vértices se pueden ordenar de la forma $V(C) = \{v_1, v_2, v_3, \dots, v_m\}$ y cuyos lados son $\{v_1v_2, v_2v_3, v_3v_4, \dots, v_{m - 1}v_m, v_mv_1\}$. Es decir, los vértices pueden disponerse de manera *circular*: cada vértice es adyacente únicamente a sus consecutivos en el círculo.

---
Anteriormente pensamos en caminos *dentro* de un grafo, lo cual nos lleva de inmediato al concepto de *subgrafo*. Un **subgrafo** de un grafo $G$ es un grafo $H$ tal que $V(H)\subseteq V(G)$, $E(H)\subseteq E(G)$ y $f_H = {f_G}_{|E(H)}$. 

Otro concepto importante es el de grafo *completo*. $G$ es un grafo **completo** si $V(G)$ es un clique. Por ejemplo, un triángulo, en el sentido del problema 2, es un grafo completo. Ahora podemos plantear de otra manera ese problema:

> ¿Es cierto que si coloreamos los lados de un grafo completo con $6$ vértices usando dos colores, tendremos un subgrafo completo con $3$ vértices cuyos lados tienen todos el mismo color? 

Volviendo al problema 4, necesitamos que dadas dos ciudades exista efectivamente un camino cuyos extremos sean las ciudades dadas. Un grafo $G$ es **conexo** si dados $v$, $w$ existe un camino $P$ que es subgrafo de $G$ y tiene como extremos a $v$ y $w$. # <span style="color:#8B0000;font-family:Papirus">Grado de un vértice</span> 

1. ¿Cómo se relaciona la cantidad de lados de un grafo con la cantidad de vértices?
2. ¿Cuántos subgrafos con características dadas, tiene un grafo?
3. ¿Cuántas clases de isomorfiso tiene un grafo dado?

Antes de analizar preguntas como estas, recordemos algunos conceptos y notación e incluyamos algunos más:
- El **grado** de $v\in V(G)$ es la cantidad de lados de $G$ incidentes con $v$ (si $G$ tiene bucles, éstos contribuyen al grado en $2$). Lo denotamos $d_G(v)$ o $d(v)$.
- $\Delta(G)$ es el máximo de los grados en $G$.
- $\delta(G)$ es el mínimo de los grados en $G$.
- Si $\Delta(G) = \delta(G) = k$, decimos que $G$ es **$k$-regular**.
- La **vecindad** de $v\in V(G)$ es el conjunto $N_G(v)$ de los vecinos de $v$ en $G$.
- El **orden** de $G$ es $n(G) = |V(G)|$.
- El **tamaño** de $G$ es $e(G) = |E(G)|$.
- Dado $n\in \mathbb{N}$, denotamos $[n] = \{1, 2, \dots, n\}$. 

Si calculamos la suma de los grados de los vértices de $G$, fíjate que cada lado se cuenta dos veces. Por lo tanto
$$ $$
$$\sum_{v\in V(G)}d_G(v) = 2e(G)$$
$$ $$
este resultado se conoce como el primer teorema de la Teoría de Grafos, o el Teorema de los estrechones de manos.

Podemos sacar varias conclusiones a partir del Teorema anterior, para un grafo $G$:
1. El promedio de los grados es $\frac{2e(G)}{n(G)}$, por lo tanto:
$$ $$
$$\delta(G) \leq \frac{2e(G)}{n(G)} \leq \Delta(G)$$
$$ $$
2. $G$ tiene una cantidad par de vértices con grado impar.
3. Si $G$ es $k$-regular entonces $e(G) = kn(G)/2$.
4. Si $n(G)$ es impar y $G$ es $k$-regular, entonces $k$ no puede ser impar.
