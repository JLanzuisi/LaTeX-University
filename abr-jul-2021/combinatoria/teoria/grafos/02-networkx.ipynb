```python
import networkx as nx 
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
G = nx.Graph()
```

```python
G.order()
```

```python
G.size()
```

```python
vertices_G = ['a', 'b', 'c', 'd', 'e']
G.add_nodes_from(vertices_G)
```

```python
G.order()
```

```python
lados_G = [('a', 'b'), ('a', 'c'), ('b', 'd'), ('c', 'd')]
G.add_edges_from(lados_G)
```

```python
G.size()
```

```python
nx.degree(G, 'a')
```

```python
list(G.degree())
```

```python
G.add_edge('e', 'b')
```

```python
G.size()
```

```python
list(G.degree())
```

```python
nx.draw(G, node_color='blue', node_size=50)
```

```python
ubica = {'a': (0, 0), 'b': (1, 0), 'c': (1, 1), 'd': (0, 1), 'e': (-1, -1)}
nx.draw(G, pos=ubica, with_labels=True, node_color='yellow')
```

```python
G.remove_edge('a', 'c')
nx.draw(G, pos=ubica, with_labels=True, node_color='yellow')
```

```python
G.size()
```

```python
G.remove_node('d')
nx.draw(G, pos=ubica, with_labels=True, node_color='yellow')
```

```python
list(G.nodes())
```

```python
list(G.edges())
```

```python
K = nx.complete_graph(30)
nx.draw_circular(K, with_labels=True, node_color='yellow')
```

```python
C = nx.cycle_graph(17)
nx.draw_circular(C, with_labels=True, node_color='yellow')
```

```python
P = nx.path_graph(12)
nx.draw(P, with_labels=True, node_color='yellow')
```

```python
B = nx.complete_bipartite_graph(4, 9)
nx.draw(B, with_labels=True, node_color='yellow')
```

```python
X, Y = nx.bipartite.sets(B)
```

```python
X
```

```python
Y
```

```python
ubica_B = nx.bipartite_layout(B, list(X))
nx.draw(B, pos=ubica_B, with_labels=True, node_color='yellow')
```

```python
S = nx.star_graph(12)
nx.draw(S, with_labels=True, node_color='yellow')
```

```python
L = nx.lollipop_graph(6, 3)
nx.draw(L, with_labels=True, node_color='yellow')
```
