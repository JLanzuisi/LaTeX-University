# <span style="color:#8B0000;font-family:Papirus">Árboles</span>

Antes vimos aplicaciones de los grafos en las que se buscan recorridos cerrados o ciclos con ciertas características. Ahora veremos otras situaciones en las que ocurre lo contrario: los ciclos están prohibidos. Por ejemplo, en un sistema de archivos organizado de forma jerárquica: una carpeta que contiene carpetas y otro tipo de archivos, cada una de las subcarpetas también contiene carpetas y archivos, así sucesivamente. Se trata de una estructura tipo árbol. 

Además de no tener ciclos, también queremos que a partir de un vértice sea posible alcanzar cualquier otro por medio de un camino. Es decir, que el grafo sea *conexo*. Así que un grafo $T$ es un árbol si satisface
> **A:** $G$ es conexo y no tiene ciclos.

Supongamo que $T$ es un árbol con al menos dos vértices.
1. $T$ tiene al menos un lado.
2. Sea $P$ un camino maximal en $T$.
3. Los extremos de $T$ tienen grado $1$.

Los vértices de grado $1$ en un árbol se llaman **hojas**. Hemos demostrado:
> <span style="color:#8B0000;font-family:Papirus">**Proposición:**</span> Todo árbol con al menos dos vértices tiene al menos dos hojas.

Si podamos un árbol (con suficiente cuidado) el resultado debería seguir siendo un árbol. Supongamos que $T$ es un árbol con $n$ vértices. ¿Qué sucede si eliminamos una hoja de $T$?
1. Sea $G = T - h$, donde $h$ es una hoja de $T$.
2. Dados $u$, $v\in V(G)$. Todo camino de $u$ a $v$ en $T$ está realmente en $G$.
3. Entonces $G$ es conexo.
4. $G$ no tiene ciclos porque $T$ tampoco los tiene.

> <span style="color:#8B0000;font-family:Papirus">**Proposición:**</span> Eliminar una hoja a un árbol con $n$ vértices resulta en un árbol con $n - 1$ vértices.

Con lo visto hasta aquí podemos concluir:
> <span style="color:#8B0000;font-family:Papirus">**Teorema:**</span> Todo árbol (finito) con más de un vértice se puede construir agregando hojas a un árbol. 

 $\quad$ 

Supongamos que $T$ es un árbol y tiene $n$ vértices. ¿Cuántos lados tiene $T$?
1. Si $n = 1$, $T\cong P_1$, no tiene lados.
2. Si $n = 2$, $T\cong P_2$, tiene un lado.
3. Si $n = 3$, $T\cong P_3$, tiene dos lados.
4. Si $n = 4$, $T\cong P_4$ o $T\cong K_{1, 3}$. Tiene tres lados.

Parece que, en general, $T$ tiene $n - 1$ lados. Supongamos que $T$ es un árbol con $n$ vértices y $n > 3$.
1. Supongamos que es el resultado es cierto para todo árbol con menos de $n$ vértices.
2. Sea $G = T - h$, donde $h$ es una hoja de $T$.
3. Por la Proposición, $G$ es un árbol con $n - 1$ vértices.
4. Por hipótesis inductiva, $G$ tiene $n - 2$ lados.
5. Como el grado de $h$ es $1$, $T$ tiene $n - 1$ lados.

> <span style="color:#8B0000;font-family:Papirus">**Lema:**</span> Todo árbol con $n$ vértices tiene $n - 1$ lados.

Observa que para que un grafo $G$ sea un árbol debe tener las dos características del enunciado **A**. Agreguemos una característica a la del Lema anterior, para árboles con $n$ vértices:
> **B:** $G$ no tiene ciclos y tiene $n - 1$ lados.

El Lema dice que **A** implica **B**. ¿Será cierto el recíporco? Supongamos **B** para un grafo $G$ con $n$ vértices.
1. Supongamos que $G_1$, $G_2$, $G_3$, $\dots$, $G_k$ son las componentes de $G$.
2. Entonces $\sum_{j = 1}^k n(G_j) = n$ porque cada vértice está en una sola componente.
3. Cada $G_j$ es conexo y no tiene ciclos, o sea es árbol.
4. Por el Lema, para cada $j$, $e(G_j) = n(G_j) - 1$.
5. Así que $n - 1 = e(G) = \sum_{j = 1}^k e(G_j) = n - k$.
6. Entonces $k = 1$, o sea, $G$ es conexo.

De manera que **A** y **B** son formas equivalentes de definir árboles. Otro par de características, para un grafo con $n$ vértices es:
> **C:** $G$ es conexo y tiene $n - 1$ lados.

Por el Lema, **A** implica **C**. No es difícil demostrar que también vale el recíproco. Veamos una característica más. Supongamos que $T$ es un árbol. Dado un par de vértices, ¿es posible que haya más de un camino que los conecte? Si es así, 
1. Elige un par de vértices y caminos $P$, $Q$ que los conecten, de manera que la suma de las longitudes de los caminos es mínima.
2. Entonces ningún vértice interno de $P$ está en $Q$, y viceversa.
3. Ahora, $P\cup Q$ es un ciclo. Esto contradice **A**.

Entonces demostramos que **A** implica el otro enunciado:
> **D:** Para cada par de vértices de $T$ existe un único camino que los conecta.

¿Será cierto que **D** implica **A**? Supongamos que $T$ es un grafo que satisface **D**.
1. Si $T$ tiene un ciclo $(v_0, v_1, v_2, \dots, v_{n - 1})$.
2. Entonces $v_{n - 1}$, $v_0$ y $v_0, v_1, v_2, \dots, v_{n - 1}$ definen caminos distintos de $v_0$ a $v_{n - 1}$. Esto contradice **D**.

Para resumir, hemos visto (salvo una de las implicaciones):
> <span style="color:#8B0000;font-family:Papirus">**Teorema:**</span> Sea $G$ un grafo con $n$ vértices. Son equivalentes:
>> 1. $G$ es un árbol.
>> 2. $G$ es conexo y $e(G) = n - 1$.
>> 3. $G$ no tiene ciclos y $e(G) = n - 1$.
>> 4. Para cada par de vértices de $G$ existe un único camino que los conecta.
